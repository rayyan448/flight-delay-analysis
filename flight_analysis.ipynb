```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv("Airline_Delay_Cause.csv")

# 2. Convert date column if not already datetime
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# 3. Extract features
df["day_of_week"] = df.index.day_name()
df["day_num"] = df.index.dayofweek             # Monday, Tuesday...
df["month"] = df.index.month_name()       # January, February...
df["hour"] = df.index.hour                # 0–23

# 4. Visualization 1: Flights by Day of Week
flights_by_day = df.groupby("day_of_week")["arr_flights"].sum().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
)

plt.figure(figsize=(8,5))
flights_by_day.plot(kind="bar")
plt.title("Flights by Day of Week")
plt.ylabel("Number of Flights")
plt.xlabel("Day of Week")
plt.xticks(rotation=45)
plt.show()

# 5. Visualization 2: Flights by Month
flights_by_month = df.groupby("month")["arr_flights"].sum()

plt.figure(figsize=(10,5))
flights_by_month.plot(kind="bar")
plt.title("Flights by Month")
plt.ylabel("Number of Flights")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()

# 6. Visualization 3: Average Delay by Hour
avg_delay_by_hour = df.groupby("hour")["arr_delay"].mean()

plt.figure(figsize=(10,5))
avg_delay_by_hour.plot(kind="line", marker="o")
plt.title("Average Arrival Delay by Hour of Day")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Hour of Day")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("Airline_Delay_Cause.csv")

# 1. Airlines with the most delays
airline_delays = df.groupby("carrier_name")["arr_del15"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=airline_delays.index, y=airline_delays.values, palette="viridis")
plt.xticks(rotation=90)
plt.title("Airlines with Most Delays")
plt.ylabel("Number of Delays")
plt.xlabel("Airlines")
plt.show()

# 2. Airports with the most delays
airport_delays = df.groupby("airport_name")["arr_del15"].sum().sort_values(ascending=False).head(15)
plt.figure(figsize=(12,6))
sns.barplot(x=airport_delays.values, y=airport_delays.index, palette="mako")
plt.title("Top 15 Airports with Most Delays")
plt.xlabel("Number of Delays")
plt.ylabel("Airports")
plt.show()

# 3. Delays by Month
monthly_delays = df.groupby("month")["arr_del15"].sum()
plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_delays.index, y=monthly_delays.values, marker="o")
plt.title("Monthly Delay Trends")
plt.xlabel("Month")
plt.ylabel("Number of Delays")
plt.grid(True)
plt.show()

# 4. Delays by Day of Week (if available)
if "day" in df.columns:
    daily_delays = df.groupby("day")["arr_del15"].sum()
    plt.figure(figsize=(10,6))
    sns.barplot(x=daily_delays.index, y=daily_delays.values, palette="coolwarm")
    plt.title("Delays by Day of Week")
    plt.xlabel("Day")
    plt.ylabel("Number of Delays")
    plt.show()

# 5. Effect of Weather on Delays
plt.figure(figsize=(10,6))
sns.scatterplot(x="weather_ct", y="arr_del15", data=df, alpha=0.5)
plt.title("Weather vs Delays")
plt.xlabel("Weather Delay Count")
plt.ylabel("Arrival Delays (15+ mins)")
plt.show()

# 6. Heatmap of different delay causes
delay_causes = df[["carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"]]
plt.figure(figsize=(8,6))
sns.heatmap(delay_causes.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Delay Causes")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("Airline_Delay_Cause.csv")

# Create DATE column from year + month and set as index
df["DATE"] = pd.to_datetime(df[["year", "month"]].assign(day=1))
df.set_index("DATE", inplace=True)

# -------------------------
# AIRLINE ANALYSIS
# -------------------------
airline_stats = df.groupby("carrier_name")[["arr_delay", "carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"]].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=airline_stats.index, y=airline_stats["arr_delay"])
plt.title("Average Arrival Delay by Airline")
plt.ylabel("Minutes")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=airline_stats.index, y=airline_stats["carrier_delay"])
plt.title("Carrier Delay by Airline")
plt.ylabel("Minutes")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=airline_stats.index, y=airline_stats["weather_delay"])
plt.title("Weather Delay by Airline")
plt.ylabel("Minutes")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.show()

# -------------------------
# AIRPORT ANALYSIS
# -------------------------
airport_stats = df.groupby("airport_name")[["arr_delay", "carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"]].mean()

top_airports = airport_stats.sort_values("arr_delay", ascending=False).head(15)

plt.figure(figsize=(14, 6))
sns.barplot(x=top_airports.index, y=top_airports["arr_delay"])
plt.title("Top 15 Airports - Average Arrival Delay")
plt.ylabel("Minutes")
plt.xlabel("Airport")
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x=top_airports.index, y=top_airports["carrier_delay"])
plt.title("Top 15 Airports - Carrier Delay")
plt.ylabel("Minutes")
plt.xlabel("Airport")
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x=top_airports.index, y=top_airports["weather_delay"])
plt.title("Top 15 Airports - Weather Delay")
plt.ylabel("Minutes")
plt.xlabel("Airport")
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x=top_airports.index, y=top_airports["late_aircraft_delay"])
plt.title("Top 15 Airports - Late Aircraft Delay")
plt.ylabel("Minutes")
plt.xlabel("Airport")
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# 🔹 1. Heatmap (Correlation of Delay Reasons)
# ==============================
plt.figure(figsize=(8,6))
delay_cols = ["weather_delay", "security_delay", "nas_delay", "late_aircraft_delay", "carrier_delay", "arr_delay"]

corr = df[delay_cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Delay Reasons & Arrival Delay")
plt.show()

# ==============================
# 🔹 2. Time Series Trend (Monthly Average Delay)
# ==============================
monthly_delay = df.groupby("month")["arr_delay"].mean()

plt.figure(figsize=(10,5))
monthly_delay.plot(marker="o")
plt.title("Average Arrival Delay by Month")
plt.xlabel("Month")
plt.ylabel("Avg Delay (minutes)")
plt.grid(True)
plt.show()

# ==============================
# 🔹 3. Time Series Trend (Daily Average Delay)
# ==============================
daily_delay = df.groupby(df.index)["arr_delay"].mean()

plt.figure(figsize=(12,5))
daily_delay.plot(alpha=0.7)
plt.title("Daily Average Arrival Delay")
plt.xlabel("Date")
plt.ylabel("Avg Delay (minutes)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Airline_Delay_Cause.csv")

# Combine year + month into a Date column (set to 1st of month)
df["DATE"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month"].astype(str) + "-01")
df.set_index("DATE", inplace=True)

# Extract weekday from index
df["DayOfWeek"] = df.index.day_name()

# Group by weekday -> average delay
avg_delay_by_day = df.groupby("DayOfWeek")["arr_delay"].mean().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
)

# Plot
plt.figure(figsize=(8,5))
avg_delay_by_day.plot(kind='bar', edgecolor='black')
plt.title("Average Flight Delay by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Average Delay (minutes)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Airline_Delay_Cause.csv")

# Create a datetime index from year + month
df["DATE"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month"].astype(str) + "-01")
df.set_index("DATE", inplace=True)

# Group by month across all years
avg_delay_by_month = df.groupby(df.index.month)["arr_delay"].mean()
avg_delay_by_month.index = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Plot
plt.figure(figsize=(10,6))
avg_delay_by_month.plot(kind='line', marker='o')
plt.title("Average Flight Delay by Month")
plt.xlabel("Month")
plt.ylabel("Average Delay (minutes)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
