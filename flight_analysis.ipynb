```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv("Airline_Delay_Cause.csv")

# 2. Convert date column if not already datetime
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# 3. Extract features
df["day_of_week"] = df.index.day_name()
df["day_num"] = df.index.dayofweek             # Monday, Tuesday...
df["month"] = df.index.month_name()       # January, February...
df["hour"] = df.index.hour                # 0–23

# 4. Visualization 1: Flights by Day of Week
flights_by_day = df.groupby("day_of_week")["arr_flights"].sum().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
)

plt.figure(figsize=(8,5))
flights_by_day.plot(kind="bar")
plt.title("Flights by Day of Week")
plt.ylabel("Number of Flights")
plt.xlabel("Day of Week")
plt.xticks(rotation=45)
plt.show()

# 5. Visualization 2: Flights by Month
flights_by_month = df.groupby("month")["arr_flights"].sum()

plt.figure(figsize=(10,5))
flights_by_month.plot(kind="bar")
plt.title("Flights by Month")
plt.ylabel("Number of Flights")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()

# 6. Visualization 3: Average Delay by Hour
avg_delay_by_hour = df.groupby("hour")["arr_delay"].mean()

plt.figure(figsize=(10,5))
avg_delay_by_hour.plot(kind="line", marker="o")
plt.title("Average Arrival Delay by Hour of Day")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Hour of Day")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Airline_Delay_Cause.csv")

# ✅ Create a datetime column (using day=1 because we only have year & month)
df["date"] = pd.to_datetime(dict(year=df["year"], month=df["month"], day=1))

# ✅ Make date the index
df = df.set_index("date")

# Now extracting features becomes super easy
df["day_of_week"] = df.index.day_name()      # Monday, Tuesday...
df["day_num"] = df.index.dayofweek           # 0=Mon, 6=Sun
df["month_name"] = df.index.month_name()     # January, Feb...
df["hour"] = df.index.hour                   # (all 0 here unless you have hh:mm:ss)

print(df.head())
```

```python
# Example plots
df.groupby(df.index.month_name()).size().plot(kind="bar", title="Flights by Month")
plt.show()

df.groupby(df.index.day_name()).size().plot(kind="bar", title="Flights by Day of Week")
plt.show()
```
